# Implement gradient descent (for simple linear regression) from scratch.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

# Data preprocessing 

```python
data = pd.read_csv('data.csv')
```

```python
X = data.iloc[:, 0]
Y =data.iloc[:, 1]
```

```python
plt.figure(figsize=(12,8))
plt.scatter(X, Y)
plt.show()
```

# Building the model 

```python
m=0
c=0

L =0.0001
epochs = 1000

n = float (len(X))

for i in range(len(X)):
    Y_pred = m*X +c
    D_m = (-2/n) * sum(X * (Y -Y_pred))
    D_c = (-2/n) * sum(Y - Y_pred)
    m = m - L*D_m
    c = c - L*D_m
print(m , c)
```

# making prediction

```python
plt.figure(figsize=(12,8))
Y_pred = m*X +c
plt.scatter(X ,Y)
plt.plot([min(X), max(X)], [min(Y_pred), max(Y_pred)], color= "red")
plt.show()
```
